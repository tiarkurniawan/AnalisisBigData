Import Library

```python
import pandas as pd
import matplotlib.pyplot as plt
```

Load Data

```python
from google.colab import drive
drive.mount('/content/drive')
```

```python
# Baca dan gabungkan file-file CSV menjadi satu DataFrame
df1 = pd.read_csv("/content/drive/MyDrive/bigdata/assessments.csv")
df2 = pd.read_csv("/content/drive/MyDrive/bigdata/courses.csv")
df3 = pd.read_csv("/content/drive/MyDrive/bigdata/studentAssessment.csv")
df4 = pd.read_csv("/content/drive/MyDrive/bigdata/studentInfo.csv")
df5 = pd.read_csv("/content/drive/MyDrive/bigdata/studentRegistration.csv")
df6 = pd.read_csv("/content/drive/MyDrive/bigdata/studentVle.csv")
df7 = pd.read_csv("/content/drive/MyDrive/bigdata/vle.csv")
```

Preprocessing

```python
# Gabungkan DataFrames menjadi satu
combined_df = pd.concat([df1, df2, df3, df4, df5, df6, df7], ignore_index=True)

# Hitung korelasi antar fitur-fitur
correlation_matrix = combined_df.corr()
```

Visualitation

```python
# Gambarkan matriks korelasi sebagai heatmap
plt.figure(figsize=(10, 8))
plt.imshow(correlation_matrix, cmap='coolwarm', interpolation='nearest')
plt.colorbar(label='Korelasi')
plt.title('Matriks Korelasi')
plt.xticks(range(len(correlation_matrix.columns)), correlation_matrix.columns, rotation=90)
plt.yticks(range(len(correlation_matrix.columns)), correlation_matrix.columns)
plt.show()
```
